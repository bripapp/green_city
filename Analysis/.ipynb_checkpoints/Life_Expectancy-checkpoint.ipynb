```python
All_City_County_Df = pd.concat([CA_SF_life_Ex_df, LA_life_Ex_df, PA_PIT_life_Ex_df], axis=0)
All_City_County_Df
```

```python
Clean_All_City_County_df = All_City_County_Df.dropna(how="any")
Clean_All_City_County_df
```

```python
# Create a boxplot to compare means
Clean_All_City_County_df.boxplot("Life Expectancy", by="State", figsize=(20, 10))

LifeExpectancy = Clean_All_City_County_df["Life Expectancy"]

fig1, ax1 = plt.subplots()
ax1.set_title('Life Expectancy of States for Year 2010 to 2015')
ax1.set_ylabel('Life Expectancy at Birth (Years)')
ax1.boxplot(LifeExpectancy)
plt.show()
quartiles = LifeExpectancy.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Life Expectancy is: {lowerq}")
print(f"The upper quartile of Life Expectancy is: {upperq}")
print(f"The interquartile range of Life Expectancy is: {iqr}")
#print(f"The the median of Life Expectancy is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers.")
print(f"Values above {upper_bound} could be outliers.")
```

```python
plt.boxplot(LifeExpectancy, Clean_All_City_County_df["State"]
                  flierprops={'marker': 'o', 'markersize': 10, 'markerfacecolor': 'fuchsia'})
plt.title("Life Expectancy of States for Year 2010 to 2015")
plt.xlabel("States")
plt.ylabel("Final Tumor Volume of Each Mouse (mm3)")
plt.grid()
```
