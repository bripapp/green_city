```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
import requests
import matplotlib.colors as mcolors
```

```python
#create bar chart showing median hh income comparison between Baton Rouge, Pittsburgh, San Francisco 
cities = ["San Francisco", "Pittsburgh", "Baton Rouge", "U.S."]
median_hh_income = [76909, 52421, 51954, 47780]
x_axis = cities

#Create a bar chart based upon the above data
plt.bar(x_axis, median_hh_income, color='forestgreen', align="center")
# Give chart labels and a title
plt.title("Median Household Income")
plt.xlabel("Cities")
plt.ylabel("Income")
```

```python
fig = plt.gcf()
fig.show()
```
